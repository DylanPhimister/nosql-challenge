```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print("Number of docs in results:", establishments.count_documents(query))
# Display the first document in the results using pprint
print("First result:")
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)
print(len(results_df))
results_df.head(10)
# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
print("Number of docs in results:", establishments.count_documents(query2))
# Display the first document in the results using pprint
print("First result:")
results2 = establishments.find(query2)
pprint(results2[0])

```

```python
# Convert the result to a Pandas DataFrame
results2_df = pd.DataFrame(results2)
print(len(results2_df))
results2_df.head(10)
# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {'RatingValue': 5,'geocode.latitude': {'$gte': latitude*(1-degree_search),'$lte': latitude*(1+degree_search)},'geocode.longitude': {'$gte': longitude*(1-degree_search),'$lte': longitude*(1+degree_search)}}
fields = {'RatingValue': 1, 'BusinessName': 1,'geocode.latitude':1,'geocode.longitude':1}
sort = [('scores.Hygiene', 1)]

results3 = list(db.establishments.find(query, fields).sort(sort))
print('Number of docs in results:', len(results3))
# Print the results
pprint(results3)
```

```python
# Convert result to Pandas DataFrame
results3_df = pd.DataFrame(results3)
print(len(results3_df))
results3_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
query_match = {'$match' : {'scores.Hygiene': 0}}
# 2. Groups the matches by Local Authority
group_match = {'$group': {'_id': {'LocalAuthorityName': '$LocalAuthorityName'}, 'count': {'$sum': 1}}}
# 3. Sorts the matches from highest to lowest
sort_match = {'$sort': {'count': -1}}
pipeline = [query_match, group_match, sort_match]
results4 = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print("Number of docs in results:", len(list(results4)))

# Print the first 10 results
pprint(results4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results4_df = pd.DataFrame(results4)
print(len(results4_df))
results4_df.head(10)
# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
```
